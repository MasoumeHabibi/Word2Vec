```python
import gensim
import pandas as pd
import os
```

```python
path = os.getcwd().replace("\\", "/")
data = pd.read_json(path + "/Cell_Phones_and_Accessories_5.json", lines = True)
data
```

```python
data.shape
```

```python
review_text = data.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
review_text.iloc[0]
```

```python
data.reviewText.loc[0]
```

```python
model = gensim.models.Word2Vec(window = 10, min_count = 2, workers = 4)
```

```python
model.model_trimmed_post_training = False
model.compute_loss = False
```

```python
model.build_vocab(review_text, progress_per = 1000)
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs = 15)
```

```python
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```

```python
model.wv.most_similar("bad")
```
